# Web Scraping

```python
import pandas as pd
```

Reading data from Wikipedia

```python
# Data about Cricket World cup
URL = "https://en.wikipedia.org/wiki/Cricket_World_Cup"
tables = pd.read_html(URL)
print("There are : ",len(tables)," tables")
print("Take look at table 0")
print(tables[7])
```

Read from particular table by matching (regex). In the wiki article we point out to this table with "Statistics" keyword.

![image](https://user-images.githubusercontent.com/51282928/126101567-977b896a-f2f2-4c17-90b2-51ffabb68711.png)

```python
tables = pd.read_html(URL, match="Statistics")
print("There are : ",len(tables)," tables")
print("Take look at table 0")
df = tables[0]

df.head()
```

```python
df['Statistics']
```

```python
df['Statistics', 'Win%*']
```
